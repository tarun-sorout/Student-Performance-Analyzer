```python
df['Total'] = df[['Math','Science','English','Computer','Social_Studies']].sum(axis=1)

df.head()
```

```python
df['Average'] = df['Total']/5

df.head()
```

```python
top_10 = df.sort_values(by='Average',ascending=False)

top_10.head(10)
```

```python
def grade(avg):
    if avg >= 90:
        return 'A'
    elif avg >= 80:
        return 'B'
    elif avg >= 70:
        return 'C'
    elif avg >= 60:
        return 'D'
    else:
        return 'F'
```

```python
df['Grade'] = df['Average'].apply(grade)

df.head()
```

```python
df['Grade'].value_counts()
```
